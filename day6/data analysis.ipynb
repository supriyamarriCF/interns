Question 7 – Filtering
Use the DataFrame from Question 6.
Tasks:
Salary > 50000
Department = IT
Age between 25 and 30
Display only Name and Salary
Employees from HR with salary greater than 45000

```python
import pandas as pd
data = {
    "name": ["Asha", "Rahul", "Priya", "Kiran", "Ravi"],
    "age": [23, 28, 30, 35, 27],
    "department": ["IT", "HR", "IT", "HR", "Finance"],
    "salary": [40000, 50000, 60000, 55000, 45000]
}
df = pd.DataFrame(data)
print(df)
print(df[df["salary"] > 50000])
print(df[df["department"] == "IT"])
print(df[(df["age"] >= 25) & (df["age"] <= 30)])
print(df[["name", "salary"]])
print(df[(df["department"] == "HR") & (df["salary"] > 45000)])
```

Question 8 – Reading Data
Create a DataFrame and save it as:
CSV
Excel
JSON
Read all three files back into Pandas and print them.

```python
import pandas as pd

# Step 1: Create the DataFrame
data = {
    "Name": ["Ravi", "Sita", "Arun", "John"],
    "Age": [25, 28, 30, 35],
    "Department": ["IT", "HR", "Finance", "IT"],
    "Salary": [50000, 45000, 60000, 70000]
}

df = pd.DataFrame(data)
# Step 2: Save the DataFrame
df.to_csv("employees.csv", index=False)
df.to_excel("employees.xlsx",index=False)
df.to_json("employees.json", orient="records")

print("Files saved successfully.\n")
# Step 3: Read the CSV file
csv_data = pd.read_csv("employees.csv")
print("CSV File")
print(csv_data)

print()
# Step 4: Read the Excel file
excel_data = pd.read_excel("employees.xlsx")
print("Excel File")
print(excel_data)
print()
# Step 5: Read the JSON file
json_data = pd.read_json("employees.json")
print("JSON File")
print(json_data)
```

Question 9 – GroupBy
Using the employee DataFrame,
Find:
Average salary
Total salary
Maximum salary
Minimum salary
Employee count

```python
import pandas as pd

data = {
    "Name": ["Ravi", "Sita", "Arun", "John", "Priya", "Kiran"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT"],
    "Salary": [50000, 45000, 60000, 70000, 48000, 65000]
}
df = pd.DataFrame(data)
print(df.groupby("Department")["Salary"].mean())
print(df.groupby("Department")["Salary"].sum())
print(df.groupby("Department")["Salary"].max())
print(df.groupby("Department")["Salary"].min())
print(df.groupby("Department")["Name"].count())
```

Question 10 – Missing Values
Tasks
Count missing values.
Fill missing Age.
Fill missing Salary.
Remove rows containing missing values.
Print the final DataFrame.

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ravi", "Sita", "Arun", "John"],
    "Age": [25, np.nan, 30, 28],
    "Salary": [50000, 40000, np.nan, 45000]
}

df = pd.DataFrame(data)

print(df)

print(df.isnull().sum())

df["Age"] = df["Age"].fillna(df["Age"].mean())

df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

print(df)
```

q:13 Complete Data Cleaning
Create a dataset containing:
Missing values
Duplicate rows
Incorrect data type

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ravi", "Ravi", "Sita", "Arun"],
    "Age": ["25", "25", np.nan, "30"],
    "Department": ["IT", "IT", "HR", "Finance"]
}

df = pd.DataFrame(data)
df = df.drop_duplicates()
df["Age"] = pd.to_numeric(df["Age"])
df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df)
```

Part 4: Feature Engineering 
Question 14 – Student Report
Create:
Math
Science
English
Create:
Total
Average
Result

```python
import pandas as pd

data = {
    "Student": ["Ravi", "Sita", "Arun"],
    "Math": [80, 70, 95],
    "Science": [90, 75, 88],
    "English": [85, 80, 92]
}

df = pd.DataFrame(data)

df["Total"] = df["Math"] + df["Science"] + df["English"]

df["Average"] = df["Total"] / 3

df["Result"] = df["Average"] >= 80

print(df)
```
