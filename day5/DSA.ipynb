```python
"""A company has recorded the performance scores of 10 employees.
Write a Python program to:
Store the performance scores in an array (Python list).
Display all the scores.
Find the highest performance score.
Find the lowest performance score.
Calculate the average performance score.
Count how many employees scored 80 or above.
Count how many employees scored below 50."""
a=list(map(int,input().split()))
print(a)
b=a[0]
c=a[0]
for i in range(1,len(a)):
    if a[i]>b:
        b=a[i]
print("highest score:",b)
for j in range(1,len(a)):
    if a[j]<c:
        c=a[j]
print("lowestscore:",c)
print(sum(a)/len(a))
count80 = 0
for score in a:
    if score >= 80:
        count80 += 1
print("Employees scoring 80 or above:", count80)
count50 = 0
for score in a:
    if score < 50:
        count50 += 1
print("Employees scoring below 50:", count50)


        
```

```python
#Write a Python program to find the second largest element in a given array of integers.
a = list(map(int, input("Enter the numbers: ").split()))
largest = a[0]
second_largest = a[0]
for i in range(1, len(a)):
    if a[i] > largest:
        second_largest = largest
        largest = a[i]
    elif a[i] > second_largest and a[i] != largest:
        second_largest = a[i]
print(largest)
print(second_largest)
```

```python
# double linked list
# Create Node
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
# Create nodes
node1 = Node(10)
node2 = Node(20)
node3 = Node(30)
node4 = Node(40)
# Connect nodes
node1.next = node2
node2.prev = node1
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3
# Display forward
current = node1

while current is not None:
    print(current.data)
    current = current.next
```

```python
#Given an array of integers, count the frequency of each element using a Hash Map
a = list(map(int, input().split()))
c = {}
for i in a:
    if i not in c:
        c[i] = 1
    else:
        c[i] += 1
for key, value in c.items():
    print(key, value)
```

```python
#Given an array of integers, print all elements that appear more than once.
a = list(map(int, input().split()))
c = {}
for i in a:
    if i not in c:
        c[i] = 1
    else:
        c[i] += 1
for j,n in c.items():
    if n>1:
        print(j,n)
```

```python
#linear search
a = list(map(int, input().split()))
target = int(input("Enter element to search: "))
found = False
for i in range(len(a)):
    if a[i] == target:
        print("Element found at index", i)
        found = True
        break
if not found:
    print("Element not found")
```

```python
#binary search
a = list(map(int, input().split()))
target = int(input("Enter element to search: "))
left = 0
right = len(a) - 1
while left <= right:
    mid = (left + right) // 2
    if a[mid] == target:
        print("Element found ", mid)
        break
    elif target < a[mid]:
        right = mid - 1
    else:
        left = mid + 1
else:
    print("Element not found")
```

```python
#bubble sort
a = list(map(int, input().split()))
n = len(a)
for i in range(n):
    for j in range(n - i - 1):
        if a[j] > a[j + 1]:
            a[j], a[j + 1] = a[j + 1], a[j]
print(a)
```

```python
a = list(map(int, input().split()))
n = len(a)
for i in range(n):
    min_index = i
    for j in range(i + 1, n):
        if a[j] < a[min_index]:
            min_index = j
    a[i], a[min_index] = a[min_index], a[i]

print(a)
```

```python
a = list(map(int, input().split()))
n = len(a)
for i in range(1, n):
    key = a[i]
    j = i - 1
    while j >= 0 and a[j] > key:
        a[j + 1] = a[j]
        j -= 1
    a[j + 1] = key
print(a)
```
