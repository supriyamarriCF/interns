```python
#Create a function that checks whether a number is an Armstrong number.
def armstrong(a):
    b=a
    sum=0
    c=len(str(a))
    while a>0:
        digit=a%10
        sum+=digit**c
        a=a//10
    if sum==b:
        print(b,"is a armstrong number")
    else:
        print(b,"is not a armstrong number")
h=int(input("enter a number: "))
armstrong(h)
    
```

```python
#Create a function that counts vowels in a string.
def vowels(a):
    count=0
    d="AEIOUaeiou"
    for i in a:
        if i in d:
            count+=1
    print(count)
letter=input("enter a string:")
vowels(letter)
```

```python
#Create a function that validates a password using rules such as minimum length, uppercase, lowercase, digit, and special character.
def password1(password):
    if len(password) < 8:
        print("Password must be at least 8 characters long.")
    upper = False
    lower = False
    digit = False
    special = False
    special_chars = "!@#$%^&*()-_=+[]{}|\\:;\"'<>,.?/"
    for ch in password:
        if ch.isupper():
            upper = True
        elif ch.islower():
            lower = True
        elif ch.isdigit():
            digit = True
        elif ch in special_chars:
            special = True

    if upper and lower and digit and special:
        return "Password is valid."
    else:
        return "Password is invalid."
password = input("Enter your password: ")
password1(password)
```
