```python
import json
import logging
import requests
from abc import ABC, abstractmethod

# Logging
logging.basicConfig(
    filename="student.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)


# Decorator
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"\nExecuting: {func.__name__}")
        logging.info(f"Function executed: {func.__name__}")
        return func(*args, **kwargs)
    return wrapper


# Abstraction
class Person(ABC):

    def __init__(self, name):
        self.name = name

    @abstractmethod
    def display(self):
        pass
# Inheritance
class Student(Person):
    def __init__(self, student_id, name, age):
        super().__init__(name)
        self.student_id = student_id
        self.age = age
        # Encapsulation
        self.__marks = 0

    def set_marks(self, marks):
        self.__marks = marks

    def get_marks(self):
        return self.__marks

    # Polymorphism
    def display(self):
        print(
            f"ID: {self.student_id} | "
            f"Name: {self.name} | "
            f"Age: {self.age} | "
            f"Marks: {self.__marks}"
        )


class Teacher(Person):

    def display(self):
        print(f"Teacher Name: {self.name}")

class StudentManager:

    def __init__(self):
        self.students = []

    @logger
    def add_student(self):

        student_id = int(input("Enter Student ID: "))
        name = input("Enter Name: ")
        age = int(input("Enter Age: "))
        marks = int(input("Enter Marks: "))

        student = Student(student_id, name, age)
        student.set_marks(marks)

        self.students.append(student)

        logging.info(f"Student Added: {name}")

        print("Student Added Successfully")

    @logger
    def view_students(self):

        if not self.students:
            print("No students available")
            return

        for student in self.students:
            student.display()

    @logger
    def save_to_json(self):

        data = []

        for s in self.students:
            data.append({
                "id": s.student_id,
                "name": s.name,
                "age": s.age,
                "marks": s.get_marks()
            })

        # Context Manager
        with open("students.json", "w") as file:
            json.dump(data, file, indent=4)

        print("Data saved successfully")

    @logger
    def load_from_json(self):

        try:
            with open("students.json", "r") as file:
                data = json.load(file)

            print("\nStudents Loaded From JSON")

            for student in data:
                print(student)

        except FileNotFoundError:
            print("students.json not found")

    # Generator
    def generate_students(self):

        for student in self.students:
            yield student

    @logger
    def generator_demo(self):

        print("\nGenerator Output")

        for student in self.generate_students():
            print(student.name)

    @logger
    def iterator_demo(self):

        if not self.students:
            print("No students available")
            return

        iterator = iter(self.students)

        print("\nIterator Output")

        try:
            while True:
                student = next(iterator)
                print(student.name)

        except StopIteration:
            print("End of Iterator")

    @logger
    def fetch_api_data(self):

        url = "https://jsonplaceholder.typicode.com/users"

        response = requests.get(url)

        if response.status_code == 200:

            users = response.json()

            print("\nUsers From API\n")

            for user in users[:5]:
                print(user["name"])

        else:
            print("Failed to fetch data")


# Main Program

manager = StudentManager()

while True:

    print("\n===== 7STUDENT MANAGEMENT SYSTEM =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Save To JSON")
    print("4. Load From JSON")
    print("5. Generator Demo")
    print("6. Iterator Demo")
    print("7. Fetch API Data")
    print("8. Polymorphism Demo")
    print("9. Exit")

    choice = input("\nEnter Choice: ")

    if choice == "1":
        manager.add_student()

    elif choice == "2":
        manager.view_students()

    elif choice == "3":
        manager.save_to_json()

    elif choice == "4":
        manager.load_from_json()

    elif choice == "5":
        manager.generator_demo()

    elif choice == "6":
        manager.iterator_demo()

    elif choice == "7":
        manager.fetch_api_data()

    elif choice == "8":

        people = [
            Student(1, "Supriya", 22),
            Teacher("Ramesh")
        ]
        print("\nPolymorphism Demo")
        for person in people:
            person.display()

    elif choice == "9":
        print("Thank You")
        break
    else:
        print("Invalid Choice")
```
