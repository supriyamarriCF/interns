```python
# Probability of getting Head
total_outcomes = 2
favourable_outcomes = 1
probability = favourable_outcomes / total_outcomes
print("Probability of Head:", probability)
```

```python
# Probability of getting an even number
dice = [1, 2, 3, 4, 5, 6]
even = 0
for i in dice:
    if i % 2 == 0:
        even += 1
probability = even / len(dice)
print("Probability of getting an even number:", probability)
```

```python
# Probability of selecting a blue ball

red_balls = 10
blue_balls = 5
total_balls = red_balls + blue_balls
probability = blue_balls / total_balls

print("Probability of selecting a blue ball:", probability)
```

```python
import numpy as np
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
print(np.corrcoef(x, y))
```

```python
import numpy as np
hours = np.array([1, 2, 3, 4, 5])
marks = np.array([40, 50, 60, 70, 80])
corr = np.corrcoef(hours, marks)[0][1]
print("Correlation:", corr)
if corr > 0:
    print("Positive Correlation")
elif corr < 0:
    print("Negative Correlation")
else:
    print("No Correlation")
```
