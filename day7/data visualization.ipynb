Question 3: Monthly Sales
Problem Statement:
A shop recorded its sales for six months.
Month	Sales
Jan	120
Feb	150
Mar	170
Apr	160
May	190
Jun	220
Create a line chart showing the monthly sales.
Requirements:
Add a title.
Label the X-axis as Months.
Label the Y-axis as Sales.

```python
import matplotlib.pyplot as plt
month = ["jan", "feb", "mar", "apr", "may", "jun"]
sales = [120, 150, 170, 160, 190, 220]
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.plot(month, sales)
plt.show()
```

Question 7: Scatter Plot
Problem Statement:
The number of study hours and marks obtained by students are:
Hours	Marks
1	30
2	40
3	50
4	70
5	90
Create a scatter plot showing the relationship between study hours and marks

```python
import matplotlib.pyplot as plt
hours=[1,2,3,4,5]
marks=[30,40,50,70,90]
plt.title("Scatter Plot")
plt.xlabel("hours")
plt.ylabel("marks")
plt.scatter(hours,marks)
plt.show()
```

```python
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
average_marks = [65, 68, 72, 75, 80, 85]

subjects = ["Math", "Science", "English", "Computer"]
subject_average = [78, 85, 72, 90]

activities = ["Sports", "Music", "Dance", "Art"]
students = [40, 25, 20, 15]

marks = [45, 50, 55, 60, 62, 65, 70, 72, 75, 80, 85, 90, 95]

study_hours = [1, 2, 3, 4, 5, 6]
exam_marks = [30, 45, 55, 70, 85, 95]
plt.plot(months, average_marks, marker="o")
plt.title("Average Marks by Month")
plt.xlabel("Months")
plt.ylabel("Average Marks")
plt.show()
plt.bar(subjects, subject_average)
plt.title("Average Marks by Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()
plt.pie(students, labels=activities, autopct="%1.1f%%")
plt.title("Student Participation in Activities")
plt.show()
plt.hist(marks, bins=5)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
plt.scatter(study_hours, exam_marks)
plt.title("Study Hours vs Exam Marks")
plt.xlabel("Study Hours")
plt.ylabel("Exam Marks")
plt.show()
```
